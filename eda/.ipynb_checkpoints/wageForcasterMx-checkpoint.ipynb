# Read file

```python
df = pd.read_csv(RAW_DATA_FILE, encoding = "ISO-8859-1")
```

```python
df.head(150)
```

# Clean data

```python
# rename columns
df = df.rename(columns= {
    'Periodo': 'period', 
    'Entidad_Federativa': 'state',
    'Sexo': 'sex',
    'Grupo_edad': 'age',
    'Nivel_ingreso': 'level',
    'Numero_personas': 'num_people'
})
```

```python
df
```

```python
# clean data
df['age'] = df['age'].str.replace(' A¥OS', '')
df['age'] = df['age'].str.replace(' ', '')
df['age'] = df['age'].str.replace('A', '-')
df['age'] = df['age'].str.replace('M-S', 'MAS')
df['age'] = df['age'].str.replace('Y', '-')
df['age'] = df['age'].str.replace('NOESPECIFIC-DO', '0')
```

```python
df.head(900)
```

# Change date

```python
# change date
```

```python
df['period'] = df['period'].astype(str)
df['year'] = df['period'].str.slice(stop=4)
df['month'] = df['period'].str.slice(start=4, stop=6)
```

```python
df
```

# Change level

```python
# change level
df['level'] = df['level'].str.replace('Menos de 1 s.m.', '-1')
df['level'] = df['level'].str.replace('1 salario m¡nimo', '1')
df['level'] = df['level'].str.replace('M\xa0s de 1 hasta 2 s.m.', '1')
df['level'] = df['level'].str.replace('M\xa0s de 2 hasta 3 s.m.', '2-3')
df['level'] = df['level'].str.replace('M\xa0s de 3 hasta 5 s.m.', '3-5')
df['level'] = df['level'].str.replace('M\xa0s de 5 hasta 10 s.m.', '5-10')
df['level'] = df['level'].str.replace('M s de 10 s.m.', '10+')
df['level'] = df['level'].str.replace('M\xa0s de 10 s.m.', '10+')
df['level'] = df['level'].str.replace('No recibe ingresos', '0')
```

```python
df.head(200)
```

```python
df.drop(df[ df['level'] == 'No especificado' ].index , inplace=True)
```

```python
df.head(250)
```

# Delete 2019

```python
df.drop(df[df['year'] == '2019'].index , inplace=True)
```
